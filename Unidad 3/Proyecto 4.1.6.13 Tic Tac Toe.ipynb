# Tic Tac Toe (Juego del Gato)

```python
from random import randrange

def DisplayBoard(board):
    
    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[0][0]  + '     |     ' +  board[0][1]  + '     |    ' +  board[0][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[1][0]  + '     |     ' +  board[1][1]  + '     |    ' +  board[1][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')
    print('|           |           |          |')
    print('|     ' +  board[2][0]  + '     |     ' +  board[2][1]  + '     |    ' +  board[2][2]  + '     |')
    print('|           |           |          |')
    print('+-----------+-----------+----------+')

def EnterMove(board):
    while True:
        move = int(input("Por favor ingrese un número dentro del rango de casillas del 1 - 9: "))
        if move < 1 or move > 9:
            print("Por favor ingrese un número en el rango del 1 hasta el 9: ")
            continue
        elif str(move) not in board[0] and str(move) not in board[1] and str(move) not in board[2]:
            print("Lo siento, esa casilla ya está ocupada. Por favor selecciona otra")
            continue
            
        for fila in range (0,3):
            for columna in range (0,3):
                if board[fila][columna] == str(move):
                    board[fila][columna] = 'O'
        break
        
def MakeListOfFreeFields(board):
    global casillas_disponibles
    casillas_disponibles = [ ]
    
    for fila in range (0,3):
        for columna in range (0,3):
            if board[fila][columna] == 'X' or board[fila][columna] == 'O':
                pass
            else:
                casillas_disponibles.append((([fila], [columna])))
        
def VictoryFor(board, sign):
    if sign == 'O':
        print("Comprobando si eres el ganador")
    else:
        print("Comprobando si la computadora ha ganado el juego")
    if board[0][0] == sign and board[0][1] == sign and board[0][2] == sign:
        return True
    elif board[1][0] == sign and board[1][1] == sign and board[1][2] == sign:
        return True
    elif board[2][0] == sign and board[2][1] == sign and board[2][2] == sign:
        return True
    elif board[0][0] == sign and board[1][0] == sign and board[2][0] == sign:
        return True
    elif board[0][1] == sign and board[1][1] == sign and board[2][1] == sign:
        return True
    elif board[0][2] == sign and board[1][2] == sign and board[2][2] == sign:
        return True
    elif board[0][0] == sign and board[1][1] == sign and board[2][2] == sign:
        return True
    elif board[2][0] == sign and board[1][1] == sign and board[0][2] == sign:
        return True
    else:
        print("¡Aún no tenemos un ganador!")
        
def DrawMove(board):
    
    while True:
        fila = randrange(3)
        columna = randrange(3)
        
        if ([fila], [columna]) not in casillas_disponibles:
            continue
        else: 
            board[fila][columna] = 'X'
            return
        
board = [['1', '2', '3'],['4', 'X', '6'],['7', '8', '9']]
nummovimientos = 1
jugador = 'O'
computadora = 'X'

print("¡Bienvenido al juego del Gato!")
print("Aquí está el estatus actual del tablero del juego : ")
DisplayBoard(board)
print()


# El siguiente código cubre el turno del usuario en cada ronda.

while nummovimientos < 9:
    EnterMove(board)
    nummovimientos += 1
    DisplayBoard(board)
    
    if VictoryFor(board, jugador) == True:
        print("¡Impresionante, has derrotado a la computadora!")
        break
    else: 
        print("Aquí estan las casillas vacías")
        MakeListOfFreeFields(board)
        print()
        DisplayBoard(board)
    
# El siguiente código cubre el turno de la computadora en cada ronda.

    print()
    print("Ahora es el turno de la computadora")
    DrawMove(board)
    nummovimientos += 1
    DisplayBoard(board)
    print()
    
    if VictoryFor(board, computadora) == True:
        print ("¡Has perdido! La computadora te ganó. Suerte para la próxima")
        break
    else: 
        print("Aquí estan las casillas vacías")
        MakeListOfFreeFields(board)
        print()
        DisplayBoard(board)
else:
    print("¡Tenemos un empate! Nadie ha ganado:(")

print("Fin del juego\n", "Gracias por jugar\n", "Nos vemos pronto")
```
